```python
import numpy as np
import pandas as pd
import tensorflow
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv("C:/Users/Casper/OneDrive/Masaüstü/ANN/titanic.csv")

#separating data into training and test sets
train, test = train_test_split(df, test_size = 0.2)

#remove unneeded columns and rows from training and test data sets
train = train.drop(['Name', 'Cabin', 'Ticket', 'PassengerId'], axis=1)
train = train.dropna()
test = test.drop(['Name', 'Cabin', 'Ticket', 'PassengerId'], axis=1)
test = test.dropna()

#convert the sex column to numeric values
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male':0, 'female':1})

#convert embarked column to numeric values
train['Embarked'] = train['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test['Embarked'] = test['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
X_train = train.drop(['Survived'], axis=1)
X_train = X_train.values
X_test = test.drop(['Survived'], axis=1)
X_test = X_test.values

#detach the 'Survived' column as a target variable
y_train = train['Survived']
y_test = test['Survived']
```

```python
# convert target variables to one-hot encoding format
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
nb_classes = 2    #number of output classes(survive and die)
nb_epoch = 100    #epoch number
batch_size = 32   #number of data samples to be used in each training step
```

```python
#defining the artificial neural network model
model = Sequential()

#input Layer
model.add(Input(shape=(X_train.shape[1],)))

#first hidden layer
model.add(Dense(32))
model.add(Activation('relu'))

#second hidden layer
model.add(Dense(64))
model.add(Activation('relu'))

#third hidden layer
model.add(Dense(128))
model.add(Activation('relu'))

#fourth hidden layer
model.add(Dense(256))
model.add(Activation('relu'))

#output layer
model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
#optimizer is used to update the weights of the model, metrics ensures model accuracy during training and testing processes
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#training the model with training data
model.fit(X_train, y_train, batch_size=batch_size, epochs=nb_epoch, validation_data=(X_test, y_test), verbose=2)
```

```python
#making predictions on test data
Y_predict = model.predict(X_test)

#thresholding
Y_predict = (Y_predict > 0.5)
```

```python
from sklearn.metrics import accuracy_score

#calculating the accuracy of predictions
print(f"Success rate is %{accuracy_score(y_test, Y_predict)*100}")
```
